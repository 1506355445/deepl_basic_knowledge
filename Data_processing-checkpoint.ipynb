```python
import numpy as np
from sklearn.impute import SimpleImputer
```

### 创建一个缺失值填充器，用 mean（均值）策略填充 np.nan。
### 拟合数据（计算每列的均值）
### 第一列：忽略 nan，计算均值 (1 + 7) / 2 = 4.0。
### 第二列：直接计算均值 (2 + 3 + 6) / 3 = 3.666...。

```python
imp = SimpleImputer(missing_values=np.nan, strategy='mean')
imp.fit([[1, 2], [np.nan, 3], [7, 6]])
X = [[np.nan, 2], [6, np.nan], [7, 6]]
print(imp.transform(X))
```

#### 首先我们生成随机数据并可视化，代码如下：



```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.datasets import make_blobs
```

```python
# 生成样本数据
X, y = make_blobs(n_samples=10000, n_features=3, 
                  centers=[[3,3,3], [0,0,0], [1,1,1], [2,2,2]],
                  cluster_std=[0.2, 0.1, 0.2, 0.2], random_state=9)

# 创建3D画布
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')  # 正确创建3D子图
ax.view_init(elev=30, azim=20)  # 设置视角

# 绘制三维散点图（正确传入x, y, z）
ax.scatter(X[:, 0], X[:, 1], X[:, 2], marker='o', s=10)  # 显式设置合理点大小

# 可选：设置坐标轴标签
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=3)
pca.fit(X)

print(pca.explained_variance_ratio_)  
# [0.98318212 0.00850037 0.00831751]
print(pca.explained_variance_)  
# [3.78521638 0.03272613 0.03202212]
```

####  可以看出投影后三个特征维度的方差比例大约为98.3%：0.8%：0.8%。投影后第一个特征占了绝大多数的主成分比例。

#### 现在我们来进行降维，从三维降到二维，代码如下：

```python
pca = PCA(n_components=2)
pca.fit(X)
print(pca.explained_variance_ratio_)  
# [ 0.98318212  0.00850037]
print(pca.explained_variance_)  
# [ 3.78483785  0.03272285]
```

#### 这个结果其实可以预料，因为上面三个投影后的特征维度的方差分别为：
#### [3.78521638   0.03272613   0.03202212]，投影到二维后选择的肯定是前两个特征，而抛弃第三个特征。
#### 我们看看此时转化后的数据分布，代码如下：


```python
X_new = pca.transform(X)
plt.scatter(X_new[:, 0], X_new[:, 1], marker='o')
plt.show()
```

#### 可见降维后的数据依然可以很清楚的看到我们之前三维图中的4个簇。


#### 现在我们看看不直接指定降维的维度，而指定降维后的主成分方差和比例。

```python
pca = PCA(n_components=0.95)
pca.fit(X)
print(pca.explained_variance_ratio_)  # [0.98318212]
print(pca.explained_variance_)  # [3.78521638]
print(pca.n_components_)  # 1
```

#### 可见只有第一个投影特征被保留。这也很好理解，我们的第一个主成分占投影特征的方差比例高达98%。只选择这一个特征维度便可以满足95%的阈值。我们现在选择阈值99%看看，代码如下：

```python
pca = PCA(n_components=0.99)
pca.fit(X)
print(pca.explained_variance_ratio_)  # [0.98318212 0.00850037]
print(pca.explained_variance_)  # [3.78521638 0.03272613]
print(pca.n_components_)  # 2
```

#### 这个结果也很好理解，因为我们第一个主成分占了98.3%的方差比例，第二个主成分占了0.8%的方差比例，两者一起可以满足我们的阈值。

```python
pca = PCA(n_components='mle', svd_solver='full')
pca.fit(X)
print(pca.explained_variance_ratio_) # [0.98318212]
print(pca.explained_variance_) # [3.78521638]
print(pca.n_components_) # 1
```

#### 可见由于我们的数据的第一个投影特征的方差占比高达98.3%，MLE算法只保留了我们的第一个特征。

#### mle ：最大似然估计(maximum likelihood estimator), 可以让PCA自动选择最佳的参数，缺点耗时很大
