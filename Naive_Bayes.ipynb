```python
import pandas as pd
from sklearn.naive_bayes import GaussianNB
```

```python
path = r'E:\教学\数据挖掘\iris\iris.data'  # 数据文件路径
data = pd.read_csv(path, header=None)
```

```python
data.head(5)
```

```python
x_prime = data[list(range(4))]  # 获取 '花瓣长度'、'花瓣宽度'的数据
print(x_prime)
```

```python
y = pd.Categorical(data[4]).codes
```

```python
y
```

```python
pd.DataFrame(y).value_counts()
```

```python
model = GaussianNB()
y_pred = model.fit(x_prime, y).predict(x_prime)
```

```python
print('要预测的数据真实分类: ',y)#打印要预测的数据真实分类
print('预测分类的结果',y_pred)#打印预测分类的结果
```

#### 模型性能评估

```python
print("预测的数据总数为 %d 其中错误分类共有 : %d" % (x_prime.shape[0], (y != y_pred).sum()))
```
